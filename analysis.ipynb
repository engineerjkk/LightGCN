```python
ratings = pd.read_csv(
    'raw/ml-1m/ratings.dat',
    sep='::', engine='python',
    names=['user_id', 'movie_id', '평점', '타임스탬프']
)
ratings[:5]
```

```python
movies = pd.read_csv(
    'raw/ml-1m/movies.dat',
    encoding='latin1',
    sep='::', engine='python',
    names=['movie_id', '제목', '장르']
)
movies[:5]
#movies.sample(5)
```

```python
data = pd.merge(ratings, users)
data = pd.merge(data, movies)
data[:5]
```
